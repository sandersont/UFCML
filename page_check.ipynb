```python
import requests
from bs4 import BeautifulSoup

HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                  'AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/124.0.0.0 Safari/537.36',
    'Referer': 'https://www.bestfightodds.com/',
}

session = requests.Session()
session.headers.update(HEADERS)
BASE = 'https://www.bestfightodds.com'

missing = [312, 314, 317, 318, 325]
query = 'ufc ' + ' '.join(str(n) for n in missing)
r = session.get(BASE + '/search?query=' + query.replace(' ', '+'), timeout=30)
soup = BeautifulSoup(r.text, 'lxml')

tables = soup.select('table')
if len(tables) >= 2:
    for tr in tables[1].select('tr'):
        tds = tr.select('td')
        if len(tds) >= 2:
            date_text = tds[0].get_text(strip=True)
            a = tds[1].select_one('a')
            if a and ('2025' in date_text or '2026' in date_text):
                print(f"  {date_text:20s} {a.get_text(strip=True):40s} {a.get('href','')}")
else:
    print("No table found")
    # show all event links anyway
    for a in soup.select('a[href*="/events/"]'):
        name = a.get_text(strip=True)
        if 'UFC' in name.upper():
            print(f"  {name:50s} {a.get('href','')}")
```
